I am interested in using the 'quantity' values of crops planted from this dataset, however the units aren't always the same (sometimes they are number of plants, sometimes square feet planted). I use another database of vegetables with best distance between crop infortation to convert from square feet units to number of vegetables units.

```python
plant_info = pd.read_csv(realpath('plantInfo-clean.csv'))  # plant info dataset
plant_info.loc[76, ['spaceInstructions']] = 'close together, 1 inch'
def search_number_string(string):  # function  to extract numerical values from string from https://stackoverflow.com/questions/26825729/extract-number-from-string-in-python/26825781#:~:text=To%20extract%20numeric%20values%20from,then%20find%20max%20of%20it.
    index_list = []
    del index_list[:]
    for i, x in enumerate(string):
        if x.isdigit() == True:
            index_list.append(i)
    start = index_list[0]
    end = index_list[-1] + 1
    number = string[start:end]
    return number

plantinfo = plant_info['spaceInstructions'].to_numpy()

space_info = []

for j, i in enumerate(plantinfo):
    numbarr = search_number_string(i).split()
    numbs = []
    
    for j in numbarr:   # making list of ideal plant spacing from mean of the suggested plant spacing in plant info dataset
        if j.isdigit() == True:
            numbs.append(float(j))
    space_info.append(np.mean(numbs))
    
plant_info['space_info'] = space_info
plant_info['Name']=plant_info['Name'].str.lower()    # changing the names in plant info dataset to match crop names in urban farm dataset
plant_info.Name = plant_info.Name.apply(lambda x: 'artichokes' if 'artichokes' in x else x)  
plant_info.Name = plant_info.Name.apply(lambda x: 'peppers' if 'peppers' in x else x)
plant_info.Name = plant_info.Name.apply(lambda x: 'strawberry' if 'strawberries'in x else x)
pinfname = plant_info.Name.unique()

plant_info['space_info_ft'] = plant_info['space_info']/12.0  # converting from inches to feet 
plant_info['Name'].unique()
```

```python
rowfeetmask = (crop['units']=='row feet')
veggies = crop['crop'][rowfeetmask].unique()

## Dataframe that has plant info
plant_info_mask = [True if x in veggies else False for x in (plant_info['Name'].to_numpy())]
pldict = dict(zip(veggies, plant_info['space_info_ft'][plant_info_mask]))

## converting 'row feet' unit elements in urban farm dataset to number of crops -> quantity in row feet/(space between plants + 1)
crop['quantity'][rowfeetmask] = crop['quantity'][rowfeetmask].to_numpy() / ( np.array([ pldict.get(x, np.nan) for x in crop['crop'][rowfeetmask]])) + 1
```

```python
rowfeetmask = (crop['units']=='square feet')
veggies = crop['crop'][rowfeetmask].unique()

## Dataframe that has plant info
plant_info_mask = [True if x in veggies else False for x in (plant_info['Name'].to_numpy())]
pldict = dict(zip(veggies, plant_info['space_info_ft'][plant_info_mask]))

## converting 'square feet' unit elements in urban farm dataset to number of crops -> quantity in sqr feet/(space between plants + 1) 
crop['quantity'][rowfeetmask] = (crop['quantity'][rowfeetmask].to_numpy())**0.5 / ( np.array([ pldict.get(x, np.nan) for x in crop['crop'][rowfeetmask]])) + 1
```

To access time information at a month precision I extract year and month from each recorded date

```python
def date_month_appender(df, date_key, date_format):
    date_local=df[date_key]
    date_time = pd.to_datetime(date_local, format=date_format)  #'%m/%d/%Y'
    df['date_time'] = date_time
   # df['year'] = pd.DatetimeIndex(df['year']).year
    df['month'] = pd.DatetimeIndex(df['date_time']).month
    df['month_year'] = pd.to_datetime(df['date_time']).dt.to_period('M').astype(str)
    df['year'] = pd.to_datetime(df['date_time']).dt.to_period('Y').astype(str).astype(int)
    #df['month'] = pd.to_datetime(df['date_time']).dt.to_period('m').astype(str)
    return df
```

### Target Variable 
Since some locations have contributed considerably more to the data set, the raw quantity value is not a good idicator of best vegetable to plant. Since I am interested in finding best vegetable to plant at a given location, compared to other crops planted at that location I normalized each quantity entry by total number of crops planted at that location. This effectiveley gives persentage of the total planting at any given location. 
Normalization of most grown crops per zip code:

$$ \frac{\text{Number of a crop planted}}{\text{ Total number of crops planted at zip code}}$$

```python
zip_to_quant = {}
for z in zips_unique:
    quant = np.sum(crop.loc[z, ['quantity']].to_numpy())
    zip_to_quant[z] = quant
```

```python
crop['quant_per_tot']=0
```

```python
for z in zips_unique:
    crop.loc[z, ['quant_per_tot']] = (crop.loc[z, ['quantity']].to_numpy())/zip_to_quant[z]
```

```python
crop = crop[crop['quant_per_tot']<0.9]  # cut on data from locations with only one crop
```

```python
def plant_normalized_per_zip(crop_data, zip_id):
    crop_zip = crop.loc[zip_id]
    num_plant = []
    for x in crop_names:
        s =np.sum(crop_zip[crop_zip['crop']==x]['quant_per_tot'].to_numpy())
        num_plant.append(s)
    len(num_plant), crop_names.shape


    num_names = list(range(len(num_plant)))
    plt.figure(figsize=(35, 10))
    plt.bar(num_names, num_plant, orientation = 'vertical', color = '#3d9973', edgecolor = 'k', linewidth = 1.2)

    plt.xticks(num_names, crop_names, rotation='vertical')

    plt.ylabel(f'Number of entries Planted in {zip_id}'); plt.xlabel('Plant'); plt.title('Number of entries per plant')
    plt.show()
```

```python
plant_normalized_per_zip(crop, zips_unique )
```
